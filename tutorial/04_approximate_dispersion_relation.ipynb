```python
from boiles.postprocessing.spectral import spectral_analyze, N, phi_n
from boiles.stencils import *
import numpy as np
import matplotlib.pyplot as plt
```

```python
real_ori, imag_ori = spectral_analyze(TENO5(d0=0.6, d1=0.3, d2=0.1, Cq=1, q=6), 3, [-2, -1, 0, 1, 2])
real_lin, imag_lin = spectral_analyze(TENO5(d0=0.55, d1=0.4, d2=0.05, Cq=1, q=6), 3, [-2, -1, 0, 1, 2])
real_comp, imag_comp = spectral_analyze(TENO5(d0=0.58675, d1=0.3265, d2=0.08675, Cq=5, q=7), 3, [-2, -1, 0, 1, 2])
real_iles, imag_iles = spectral_analyze(TENO5(d0=0.525, d1=0.45, d2=0.025, Cq=56000, q=13), 3, [-2, -1, 0, 1, 2])
```

```python
plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams["mathtext.fontset"] = "stix"

plt.figure(figsize=(5, 4), dpi=200)
plt.plot(phi_n, phi_n, 'k-', label="spectral", linewidth=0.5)

plt.plot(phi_n, real_ori, c="blue", linewidth=0.5)
plt.scatter(phi_n[::4], real_ori[::4], label=r"$\rm TENO5$", marker="s", facecolor='blue', edgecolor="blue", linewidth=0.5, s=20)

plt.plot(phi_n, real_lin, c="red", linewidth=0.5)
plt.scatter(phi_n[::4], real_lin[::4], label=r"$\rm TENO5_{adc}$", marker="o", facecolor='red', edgecolor="red", linewidth=0.5, s=20)

plt.plot(phi_n, real_comp, c="green", linewidth=0.5)
plt.scatter(phi_n[::4], real_comp[::4], label=r"$\rm TENO5_{comp}$", marker="^", facecolor='green', edgecolor="green", linewidth=0.5, s=20)

plt.plot(phi_n, real_iles, c="brown", linewidth=0.5)
plt.scatter(phi_n[::4], real_iles[::4], label=r"$\rm TENO5_{iles}$", marker="*", facecolor='brown', edgecolor="brown", linewidth=0.5, s=20)

plt.legend()
plt.ylim(-0.5, 2)
plt.xlim(phi_n[0], phi_n[-1])
plt.ylabel(r"$Im(\Phi(\varphi))$")
plt.xlabel(r"$\varphi$")
```
