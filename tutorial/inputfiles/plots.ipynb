```python
import matplotlib.pyplot as plt
import numpy as np
from boiles.objective.simulation1d import Simulation1D
from boiles.objective.simulation2d import Simulation2D
```

```python
case = Simulation1D("./lax/domain/data_0.140*.h5")
plt.figure(figsize=(10, 4), dpi=100)
plt.subplot(1, 2, 1)
plt.plot(case.result["x_cell_center"], case.result["density"], color="black", linewidth=0.8)
plt.xlim(0, 1)
plt.xlabel("$x$")
plt.ylabel(r"$\rho$")
plt.tight_layout()
plt.subplot(1, 2, 2)
plt.plot(case.result["x_cell_center"], case.result["velocity"], color="black", linewidth=0.8)
plt.xlim(0, 1)
plt.xlabel("$x$")
plt.ylabel(r"$u$")
plt.tight_layout()
plt.savefig("/home/yiqi/PycharmProjects/boiles/tutorial/inputfiles/pics/lax.jpg", dpi=400)
```

```python
case = Simulation2D("./2d_riemann/domain/data_1.100000.h5")
plt.figure(figsize=(8, 4), dpi=100)
plt.subplot(1, 2, 1)
plt.imshow(case.result["density"], vmin=0.15, vmax=1.7, cmap="jet", extent=[0, 1.2, 0, 1.2], origin="lower")
plt.tight_layout()
plt.subplot(1, 2, 2)
plt.contour(case.result["density"], extent=[0, 1.2, 0, 1.2], levels=np.arange(0.15, 1.70001, 0.05), colors="black", linewidths=0.4)
plt.tight_layout()
plt.savefig("/home/yiqi/PycharmProjects/boiles/tutorial/inputfiles/pics/2d_riemann.jpg", dpi=400)
```
