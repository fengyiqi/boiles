## Read 1D simulation

```python
from boiles.objective.simulation1d import Simulation1D

sod = Simulation1D(file="/media/yiqi/Fengyiqi/TUM/TENO5/0504/teno5_eta_sod_conference/presentation/sod/sod_weno5/domain/data_0.200000.h5")
print(sod.result["density"])
```

`.result` is a dictionary containing basic states

```python
import matplotlib.pyplot as plt

x = sod.result["x_cell_center"]
plt.figure(figsize=(6, 4), dpi=100)
plt.plot(x, sod.result["density"], "k-", label=r"$\rho$")
plt.plot(x, sod.result["velocity"], "b-", label=r"$u$")
plt.plot(x, sod.result["pressure"], "r-", label=r"$p$")
plt.xlim(0, 1)
plt.legend()
plt.tight_layout()
```
